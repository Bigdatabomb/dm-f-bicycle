```python

# 1. 파일 불러오기
rain = pd.read_csv("2024년 강수량 측정 일자 및 자치구 정보.csv", encoding='cp949')

# 2. 조사년월일을 datetime으로 변환
rain['조사년월일'] = pd.to_datetime(rain['조사년월일'], format='%Y%m%d')

# 3. 1월 데이터만 필터링
rain_jan = rain[rain['조사년월일'].dt.month == 1].copy()

# 4. 날짜를 'YYMMDD' 문자열로 변환 (예: 240101)
rain_jan['조사년월일'] = rain_jan['조사년월일'].dt.strftime('%y%m%d')

# 5. 저장
rain_jan.to_csv("강수량_2024_01.csv", index=False, encoding='cp949')


# 1. 파일 불러오기
months = pd.read_csv("1월 따릉이 정보.csv", encoding='cp949')
rain = pd.read_csv("강수량_2024_01.csv", encoding='cp949')

# 2. 날짜 형식 맞추기: 둘 다 'YYYYMMDD' 문자열로 통일
months['기준_날짜'] = months['기준_날짜'].astype(str).str.zfill(8)
rain['조사년월일'] = pd.to_datetime(rain['조사년월일'], format="%y%m%d").dt.strftime('%Y%m%d')

# 3. 구 이름 공백 제거 + NaN 제거
months = months.dropna(subset=['구'])
rain = rain.dropna(subset=['측정지역'])
months['구'] = months['구'].astype(str).str.strip()
rain['측정지역'] = rain['측정지역'].astype(str).str.strip()

# 4. 병합
merged = months.merge(
    rain,
    left_on=['기준_날짜', '구'],
    right_on=['조사년월일', '측정지역'],
    how='inner'
)

# 5. 정리 및 저장
merged = merged.drop(columns=['조사년월일', '측정지역'])
merged.to_csv("1월 따릉이 강수량 포함.csv", index=False, encoding='cp949')
```
